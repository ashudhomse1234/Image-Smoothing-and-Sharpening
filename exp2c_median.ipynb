```python
!pip install opencv-python-headless
from google.colab.patches import cv2_imshow # Import the necessary function from google.colab.patches
import cv2
import numpy as np

# Reading the image
image = cv2.imread('/content/download.png')

# Applying the filter
medianBlur = cv2.medianBlur(image, 9)

# Showing the image using cv2_imshow instead of cv2.imshow
cv2_imshow(image) # Display the original image
cv2_imshow(medianBlur) # Display the blurred image

cv2.waitKey()
cv2.destroyAllWindows()
```
